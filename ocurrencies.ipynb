```python
# agora finalizamos mudando o nome das colunas para os pré estabelecidos
```

```python
ocurrencies.columns = ["id", "machine_id", "material_cost", "service_cost", "total_cost", "description", "created_at"]
```

```python
list_of_twos = [2]*569

time_stopped = []
for i in range(1,570):
    time = numpy.random.choice(numpy.arange(1, 11), p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.05, 0.02, 0.02, 0.005, 0.005])
    time_stopped.append(time)
```

```python
df1 = pandas.DataFrame(numpy.array(list_of_twos), columns=['category'])
df2 = pandas.DataFrame(numpy.array(time_stopped), columns=['time_stopped'])

ocurrencies = pandas.concat([ocurrencies, df2, df1], axis = 1)
```

```python
# note que até a entrada 200 os dados não são úteis
```

```python
ocurrencies = ocurrencies.iloc[200:-1]
```

```python
ocurrencies.reset_index(inplace=True)
ocurrencies.drop(['index'],
  axis='columns', inplace=True)
```
